```python
import pandas as pd
record_origin = pd.read_csv('data/network/cnra/period_of_record.csv')
station_origin = pd.read_csv('data/network/cnra/stations.csv')
field_origin = pd.read_csv('data/network/cnra/field_results.csv')
lab_origin = pd.read_csv('data/network/cnra/lab_results.csv')
```

```python
# 只保留采样数大于100的
record_df = record_origin[record_origin['SAMPLE_COUNT'] > 100]
record_df.reset_index(drop=True, inplace=True)
# record_df.describe()
# print(len(record_origin))
print(len(record_df))

merge_df = None

for index,row in record_df.iterrows():
    # print(row['STATION_NUMBER'],row['PARAMETER'])
    sta_number = row['STATION_NUMBER']
    paramer = row['PARAMETER']
    one_df = lab_origin[(lab_origin['STATION_NUMBER'] == sta_number)
                        & (lab_origin['PARAMETER'] == paramer)]
    if len(one_df) == 0:
        print('searching another table')
        one_df = field_origin[(field_origin['STATION_NUMBER'] == sta_number)
                        & (field_origin['PARAMETER'] == paramer)]
        if len(one_df) == 0:
            continue

    # 拼接
    merge_df = pd.concat([merge_df,one_df])

    # print(len(merge_df))

    # print(index,'---')
    if index % 1000 == 0 :
        print(index)
merge_df.to_csv('data/network/cnra/middle/merge.csv')

```

```python

factors =  merge_df['PARAMETER'].unique()
for fac in factors:
    one_df = merge_df[merge_df['PARAMETER'] == fac]
    print(len(one_df),fac)

```

```python
pH_df = merge_df[merge_df['PARAMETER'] == 'pH']
print(len(pH_df))
```
